```python
import networkx as nx
G=nx.read_edgelist('facebook_combined.txt')
```

```python
from collections import deque
visited=[]
def dfs(s,d):
    visited.clear()
    stack=deque()
    stack.append(s)
    visited.append(s)
    while(stack):
        a=stack.pop()
        if(a==d):
            break
        L=list(G.adj[a])
        for i in L:
            if(i not in visited):
                stack.append(i)
                visited.append(i)
    print("Nodes traversed for finding a particular node =",len(visited))
    return
```

```python

from collections import deque
visited=[]
path=[]
pred={}
def bfs(s,d):
    visited.clear()
    path.clear()
    pred.clear()
    queue=deque()
    queue.append(s)
    visited.append(s)
    pred[s]=None
    while(queue):
        a=queue.popleft()
        #print(a,end="->")
        if(a==d):
            while a is not None:
                path.append(a)
                a = pred[a]
            path.reverse()
            break
        L=list(G.adj[a])
        for i in L:
            if(i not in visited):
                queue.append(i)
                visited.append(i)
                pred[i]=a
    print("Length of shortest path =",len(path),"\nNodes traversed for finding shortest path =",len(visited))
    return
```
